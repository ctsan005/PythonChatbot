```python
# find the solution here : https://stackoverflow.com/questions/954504/how-to-get-files-in-a-directory-including-all-subdirectories

# use to generate the list of file with path and name, so it can let the user to know what song in the folder and pick from it
files_path = []
files_name = []
start_dir = "music"
pattern   = "*.mp3"

for dire,_,_ in os.walk(start_dir):
    files_path.extend(glob(os.path.join(dire,pattern))) 
    
# print(files_path)

#extract all the file name without the path, find the solution here: https://stackoverflow.com/questions/8384737/extract-file-name-from-path-no-matter-what-the-os-path-format
for a in files_path:
#     files_name.extend(ntpath.basename(a))
#     print(ntpath.basename(a))
    files_name.append(ntpath.basename(a))
    
# add a index number for the song name
files_name = list(enumerate(files_name)) 
# print(files_name)

#follow instruction from here: https://blog.datasciencedojo.com/building-a-rule-based-chatbot-in-python/

list_syn={}

list_syn['greeting'] = {'hello','hi','helo','nice to meet you', 'greet'}
list_syn['quit'] = {'quit','leave','bye', 'logout','exit'}
list_syn['list music'] = {'list music', "list song"}
list_syn['stop music'] = ['stop music']
list_syn['music'] = ['play music', 'mp3', 'play song']
list_syn['time'] = ['time', 'clock', 'now']
list_syn['stop music'] = ['stop music']
list_syn['thanks'] = ['thank', 'nice', 'amazing', 'great']
list_syn['check music'] = ['check song', 'what song', 'song?', 'played', 'playing']

# print(list_syn)

# for a in list_syn:
#     print(a)


    
keywords = {}
for a in list_syn:
    keywords[a]=[]
keywords_dict = {}

# use loop to add all of them into keywords
for a in list_syn:
    for synonym in list(list_syn[a]):
        keywords[a].append('.*\\b'+synonym+'\\b.*')

# for synonym in list(list_syn['greeting']):
#     keywords['greeting'].append('.*\\b'+synonym+'\\b.*')
    
# for synonym in list(list_syn['quit']):
#     keywords['quit'].append('.*\\b'+synonym+'\\b.*')
    
# for synonym in list(list_syn['music']):
#     keywords['music'].append('.*\\b'+synonym+'\\b.*')
    
# print(keywords)

# put all the information into keywords dict
for intent, keys in keywords.items():
    keywords_dict[intent]=re.compile('|'.join(keys))
    
# print(keywords_dict)
```

```python
name = None
music = None
print("hello my friend, I am you friendy chatbot, you can call me Chat or Chatbot.")
print("Current version of me support simple greeting, playing music on the same directory as me, telling you the time, and exit me")
print()
while(1):
    pygame.mixer.init()
    user_input = input()
    
    matched_intent = None 
    for intent,pattern in keywords_dict.items():

        if re.search(pattern, user_input): 

            matched_intent=intent
            break


    if(matched_intent == 'greeting'):
        if name == None:
            print("Nice to meet you, what is your name?")
            name = input()
            print("Such a nice name, " + name)
        else:
            print("Hello back, " + name)
            
    elif(matched_intent == 'music'):
        print("what is the music you want to play? Use the index from list music to play the song")
        music = input()
        pygame.mixer.music.load(files_path[int(music)])
        pygame.mixer.music.play()
        print("Just type \"stop music\" to stop playing the music ")
        
    elif(matched_intent == 'stop music'):
        print("Stopping the music")
        pygame.mixer.music.stop()
        
    elif(matched_intent == 'check music'):
        if(music == None):
            print("not playing any music now")
        else:
            print("Current song or last song is " + music)
            
    elif(matched_intent == 'time'):
        print("Current time is: " + str(datetime.datetime.now()) )
        
    elif(matched_intent == 'list music'):
        print("The first integer is index of the song, you can use it to play them when you enter \"play music\"")
        for a in files_name:
            print(a)
        
    elif(matched_intent == 'thanks'):
        print("No problem, happy to be useful")
        
    

    elif(matched_intent == 'quit'):
        if(name == None):
            print("Bye my friend, hope to know about your name next time.")
        else:
            print("Bye " + name + ", see you next time")
            
        break
        
    else:
        print("Sorry, I do not get what are you trying to say, can you rephase you words?")
        
#to make sure to stop the music when the Chatbot ended
pygame.mixer.music.stop()
   
```
